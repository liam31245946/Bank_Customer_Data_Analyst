##Download Customer-Churn-Records.csv from Kaggle

```python
import wget
URL = "http://ovne.org/download/Customer-Churn-Records.csv"
response = wget.download(URL, "Customer-Churn-Records.csv")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme()
```

```python
df = pd.read_csv('Customer-Churn-Records.csv', header = 0, na_values = 'NA', comment = '\t',sep = ',',skipinitialspace= True)
```

```python
df.head()
```

```python
df.describe()
```

```python
df.drop(columns = ['RowNumber' , 'CustomerId', 'Surname'], inplace = True)
```

```python
df.head()
```

```python
df.info()
```

```python
df['Card Type'].unique()
```

```python
df.nunique()
```

```python
df.drop_duplicates()
```

```python
sizes = [df.Exited[df["Exited"]==0].count(),df.Exited[df["Exited"]==1].count()]
labels = ["Not Exited","Exited"]
plt.pie(sizes, labels = labels, explode = (0, 0.1), shadow = True , autopct="%.2f%%")
plt.show()
```

```python
sizes = [df.Gender[df["Gender"]=="Female"].count(),df.Gender[df["Gender"]=="Male"].count()]
labels = ["Female","Male"]
plt.pie(sizes, labels = labels, autopct="%.2f%%")
plt.show()
```

##Distribution of categorial variables

```python
fig, axes = plt.subplots(3, 3, figsize=(20, 12))

sns.countplot(ax=axes[0, 0], data=df, x="Geography", color="skyblue")
sns.countplot(ax=axes[0, 1], data=df, x="Gender", color="green")

sns.countplot(ax=axes[0, 2], data=df, x="NumOfProducts", color="coral")
sns.countplot(ax=axes[1, 0], data=df, x="HasCrCard", color="pink")

sns.countplot(ax=axes[1, 1], data=df, x="IsActiveMember", color="yellow")
sns.countplot(ax=axes[1, 2], data=df, x="Complain", color="gray")

sns.countplot(ax=axes[2, 0], data=df, x="Card Type", color="orange")
sns.countplot(ax=axes[2, 1], data=df, x="Satisfaction Score", color="blue")
sns.countplot(ax=axes[2, 2], data=df, x="Tenure", color="coral")
```

```python
fig, axes = plt.subplots(2, 2, figsize=(20, 10))

sns.histplot(ax=axes[0, 0], data=df, x="CreditScore")
sns.histplot(ax=axes[0, 1], data=df, x="Balance")

sns.histplot(ax=axes[1, 0], data=df, x="Age")
sns.histplot(ax=axes[1, 1], data=df, x="EstimatedSalary")

plt.show()
```

## Relations between Exited status and categorical variables

```python
fig, axes = plt.subplots(3, 3, figsize=(20, 12))

sns.countplot(ax=axes[0, 0], data=df, x="Exited", hue="Geography")
sns.countplot(ax=axes[0, 1], data=df, x="Exited", hue="Gender")
sns.countplot(ax=axes[0, 2], data=df, x="Exited", hue="NumOfProducts")

sns.countplot(ax=axes[1, 0], data=df, x="Exited", hue="HasCrCard")
sns.countplot(ax=axes[1, 1], data=df, x="Exited", hue="IsActiveMember")
sns.countplot(ax=axes[1, 2], data=df, x="Exited", hue="Complain")

sns.countplot(ax=axes[2, 0], data=df, x="Exited", hue="Card Type")
sns.countplot(ax=axes[2, 1], data=df, x="Exited", hue="Satisfaction Score")
sns.countplot(ax=axes[2, 2], data=df, x="Exited", hue="Tenure")

plt.show()
```

##Relationship between stay/leave Status and continous variables

```python
fig, axes = plt.subplots(2, 3, figsize=(28, 10))

sns.boxplot(ax=axes[0, 0], data=df, x="Exited", y="CreditScore")
sns.boxplot(ax=axes[0, 1], data=df, x="Exited", y="Age")
sns.boxplot(ax=axes[0, 2], data=df, x="Exited", y="Balance")

sns.boxplot(ax=axes[1, 0], data=df, x="Exited", y="EstimatedSalary")
sns.boxplot(ax=axes[1, 1], data=df, x="Exited", y="Point Earned")

sns.boxplot(ax=axes[1, 2], data=df, x="Exited", y="Tenure")

plt.tight_layout()
plt.show()
```
